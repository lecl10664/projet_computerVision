```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
detector = cv2.SIFT_create()
```

```python
def read_image_and_compute_keypoints(object_imgs, scene_imgs, detector):
    imgs_dict = {}

    for name_img in object_imgs:
        img_load = cv2.imread('models/' + name_img + '.jpg')
        kp, des = detector.detectAndCompute(img_load, None)
        imgs_dict[name_img] = {'kp': kp, 'des': des, 'shape': img_load.shape}

    for name_img in scene_imgs:
        img_load = cv2.imread('scenes/' + name_img + '.png')
        kp, des = detector.detectAndCompute(img_load, None)
        imgs_dict[name_img] = {'kp': kp, 'des': des, 'shape': img_load.shape}

    return imgs_dict
```

```python
def matching(descriptors_object, descriptors_scene, ratio_test=0.75, k=2):
    matcher = cv2.DescriptorMatcher_create(cv2.DescriptorMatcher_FLANNBASED)
    matches = matcher.knnMatch(descriptors_object, descriptors_scene, k)
    
    #-- store all the good matches as per Lowe's ratio test.
    good = []
    for m,n in matches:
        if m.distance < ratio_test*n.distance:
            good.append(m)
    
    return good
```

```python
def localize_object_in_scene(img_object, keypoints_object, keypoints_scene, good): #return the corners of object in the scene
    #-- Get the keypoints from the good matches
    obj = np.float32([ keypoints_object[m.queryIdx].pt for m in good ])
    scene = np.float32([ keypoints_scene[m.trainIdx].pt for m in good ])
    
    
    #-- Get the corners of object
    h,w = img_object.shape[0:2]
    obj_corners = np.float32([ [0,0],[0,h-1],[w-1,h-1],[w-1,0] ]).reshape(-1,1,2)
    
    #-- Get the homography of object/scene
    H, _ =  cv2.findHomography(obj, scene, cv2.RANSAC)
    
    #-- Get the corners of object in scene
    scene_corners = cv2.perspectiveTransform(obj_corners, H)
    
    return scene_corners
    
    
```

```python
def matching_and_localize_objects_in_scene(imgs_dict, scene_imgs, ratio_test=0.60):
    all_scene_corners = {}
    for name in object_imgs:
        object_des, object_kp = imgs_dict[name]['des'], imgs_dict[name]['kp']
        scene_des, scene_kp = imgs_dict[scene_imgs]['des'], imgs_dict[scene_imgs]['kp']
        
        #-- matching
        matcher = cv2.DescriptorMatcher_create(cv2.DescriptorMatcher_FLANNBASED)
        matches = matcher.knnMatch(object_des, scene_des, k=2)
    
        #-- store all the good matches as per Lowe's ratio test.
        good = []
        for m,n in matches:
            if m.distance < ratio_test*n.distance:
                good.append(m)
                
        #-- Get the keypoints from the good matches
        object_good_kp = np.float32([ object_kp[m.queryIdx].pt for m in good ])
        scene_good_kp = np.float32([ scene_kp[m.trainIdx].pt for m in good ])   
        
        
        #-- Get the corners of object
        h,w = imgs_dict[name]['shape'][0:2]
        object_corners = np.float32([ [0,0],[0,h-1],[w-1,h-1],[w-1,0] ]).reshape(-1,1,2)
        
        #-- Get the homography of object/scene
        H, _ =  cv2.findHomography(object_good_kp, scene_good_kp, cv2.RANSAC)

        #-- Get the corners of object in scene
        scene_corners = cv2.perspectiveTransform(object_corners, H)
        all_scene_corners[name] = scene_corners
        
    return all_scene_corners
```

```python
def drawn_object_lines_in_scene(img_scene, all_scene_corners):
    for name in object_imgs:
        scene_corners = all_scene_corners[name]
        img_scene = cv2.polylines(img_scene, [np.int32(scene_corners)], 
                                  isClosed=True, color=(0,255,0), thickness=3)
    return img_scene
```

```python
#-- bank of image
object_imgs = ['0', '1', '11', '19', '24', '25', '26']
#object_imgs = ['0', '11']
scene_imgs = ['e1', 'e2', 'e3', 'e4', 'e5']
```

```python
detector = cv2.SIFT_create()
matcher = cv2.DescriptorMatcher_create(cv2.DescriptorMatcher_FLANNBASED)
```

```python
imgs_dict = read_image_and_compute_keypoints(object_imgs, scene_imgs, detector)
```

```python
all_scene_corners = matching_and_localize_objects_in_scene(imgs_dict, 'e1', ratio_test=0.75)
```

```python
img_scene = cv2.imread('scenes/e1.png')
img_scene = drawn_object_lines_in_scene(img_scene, all_scene_corners)

plt.imshow(img_scene)
plt.show()
```
