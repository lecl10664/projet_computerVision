```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
#-- Bank of images
model_imgs = ['0', '11']
scene_imgs = ['e1']
```

```python
detector = cv2.SIFT_create()
```

```python
def read_image_and_compute_keypoints(path_file, detector):
    image = cv2.cvtColor(cv2.imread(path_file), cv2.COLOR_BGR2RGB)
    kp, des = detector.detectAndCompute(image, None) # return image, keypoints, descriptors
    return image, kp, des
```

```python
def matching(descriptors_object, descriptors_scene, ratio_test=0.75, k=2):
    matcher = cv2.DescriptorMatcher_create(cv2.DescriptorMatcher_FLANNBASED)
    matches = matcher.knnMatch(descriptors_object, descriptors_scene, k)
    
    #-- store all the good matches as per Lowe's ratio test.
    good = []
    for m,n in matches:
        if m.distance < ratio_test*n.distance:
            good.append(m)
    
    return good
```

```python
def localize_object(img_object, keypoints_object, keypoints_scene, good): #return the corners of object in the scene
    #-- Get the keypoints from the good matches
    obj = np.float32([ keypoints_object[m.queryIdx].pt for m in good ])
    scene = np.float32([ keypoints_scene[m.trainIdx].pt for m in good ])
    
    H, _ =  cv2.findHomography(obj, scene, cv2.RANSAC)
    
    #-- Get the corners of object
    h,w = img_object.shape[0:2]
    obj_corners = np.float32([ [0,0],[0,h-1],[w-1,h-1],[w-1,0] ]).reshape(-1,1,2)
    
    #-- Get the corners of object in scene
    scene_corners = cv2.perspectiveTransform(obj_corners, H)
    
    return scene_corners
    
    
```

```python
def drawn_object_lines_in_scene(img_scene, scene_corners):
    return cv2.polylines(img_scene, [np.int32(scene_corners)], isClosed=True, color=(0,255,0), thickness=3)
```

```python
img1, kp1, des1 = read_image_and_compute_keypoints(path_file='models/0.jpg', detector=detector)
img2, kp2, des2 = read_image_and_compute_keypoints(path_file='models/11.jpg', detector=detector)
img3, kp3, des3 = read_image_and_compute_keypoints(path_file='scenes/e1.png', detector=detector)
```

```python
good1 = matching(des1, des3)
good2 = matching(des2, des3)
```

```python
scene_corners1 = localize_object(img1, kp1, kp3, good1)
scene_corners2 = localize_object(img2, kp2, kp3, good2)
```

```python
img_scene = cv2.polylines(img3, [np.int32(scene_corners1)], isClosed=True, color=(0,255,0), thickness=3)
img_scene = cv2.polylines(img3, [np.int32(scene_corners2)], isClosed=True, color=(0,255,0), thickness=3)



plt.figure(figsize=(10,10)), plt.imshow(img_scene)
plt.show()
```
